```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel('city.xlsx')
```

```python
df.info()
```

```python
request_hours = df.groupby('hour',as_index=True).agg({'requests':'sum'})
```

```python
request_hours
```

```python
import matplotlib.pyplot as plt
```

```python
request_hours.plot(kind='bar')
plt.rcParams['font.sans-serif'] = ['PingFang SC', 'Hiragino Sans GB', 'Heiti SC', 'SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.xticks(rotation=0)
plt.title('各小时订单请求总量')
```
