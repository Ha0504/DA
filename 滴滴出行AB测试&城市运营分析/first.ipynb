```python
import pandas as pd
import numpy as np
```

```python
city = pd.read_excel('city.xlsx')
test = pd.read_excel('test.xlsx')
```

```python
test.head()
```

```python
test['net_gmv'] = test['gmv'] - test['coupon per trip']*test['trips']
```

```python
requests_control = test[test.group == 'control'].requests
requests_experiment = test[test.group == 'experiment'].requests
```

```python
import scipy.stats as stats
```

```python
stats.ttest_ind(requests_control, requests_experiment,equal_var=False)
```

```python
net_gmv_control = test[test.group == 'control'].net_gmv
net_gmv_experiment = test[test.group == 'experiment'].net_gmv
```

```python
stats.ttest_ind(net_gmv_control, net_gmv_experiment,equal_var=False)
```

```python
stats.levene(net_gmv_control, net_gmv_experiment, center='median')
```

用配对法

```python
daily_requests = test.pivot(index='date',columns = 'group',values = 'requests')
```

```python
daily_requests.head()
```

```python
requests_control = daily_requests['control']
requests_experiment = daily_requests['experiment']
```

```python
stats.ttest_rel(requests_control, requests_experiment)
```

```python
daily_gmv = test.pivot(index = 'date',columns='group',values = 'net_gmv')
```

```python
gmv_control = daily_gmv['control']
gmv_experiment = daily_gmv['experiment']
```

```python
stats.ttest_rel(gmv_control, gmv_experiment)
```
